### Leer Datos CSV de la carpeta "Employee Training"
1. Leer los archivos de la carpeta "employee_training"
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Leer los archivos de la carpeta "employee_training"

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, TimestampType
emp_train_schema = StructType([
    StructField("employee_id", IntegerType(), True),
    StructField("training_id", IntegerType(), True),
    StructField("completition_date", TimestampType(), True),
    StructField("status", IntegerType(), True)
])
```

```python
emp_train_df = (
  spark.read
        .option("delimiter",",")
        .schema(emp_train_schema)
        .csv("abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/employee_training/")
)
display(emp_train_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
emp_train_df.writeTo("zenviro.bronze.employee_training_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.employee_training_py;
```
