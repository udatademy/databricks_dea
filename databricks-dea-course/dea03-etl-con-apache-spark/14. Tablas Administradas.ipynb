### Tablas Administradas(Managed Tables)
1. Crear una "Tabla Administrada(Managed Table)" con Python
2. Describir(Desbribe) la "Tabla Administrada(Managed Table)"
3. Efectos de eliminar una "Tabla Administrada(Managed Table)"
4. Recuperar una "Tabla Administrada(Managed Table)"

#### 1. Crear una "Tabla Administrada(Managed Table)" con Python

```python
location_df = spark.read.format("json").load("/Volumes/zenviro/raw/operational_data/locations/")
display(location_df)
```

```python
location_df.write.format("delta").mode("overwrite").saveAsTable("zenviro.default.locations_man_python")
```

```python
display(spark.sql("SELECT * FROM zenviro.default.locations_man_python"))
```

#### 2. Describir(Desbribe) la "Tabla Administrada(Managed Table)"

```python
spark.sql("USE CATALOG zenviro;")
spark.sql("USE default;")
display(spark.sql("SHOW TABLES;"))
```

```python
display(spark.sql("DESCRIBE EXTENDED zenviro.default.locations_man_python"))
```

#### 3. Efectos de eliminar una "Tabla Administrada(Managed Table)"

```python
spark.sql("DROP TABLE IF EXISTS zenviro.default.locations_man_python")
```

```python
spark.sql("USE CATALOG zenviro;")
spark.sql("USE default;")
display(spark.sql("SHOW TABLES;"))
```

#### 4. Recuperar una "Tabla Administrada(Managed Table)"

```python
display(spark.sql("SHOW TABLES DROPPED IN zenviro.default"))
```
