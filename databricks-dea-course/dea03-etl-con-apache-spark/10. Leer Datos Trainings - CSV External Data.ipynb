### Leer Datos CSV de la carpeta "Trainings"
1. Leer todos los archivos "trainings"
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Leer todos los archivos "trainings"

```python
trainings_schema = "training_id INTEGER, training_name STRING, description STRING, provider STRING, duration_hours INTEGER"
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
trainings_schema = StructType([
    StructField("training_id", IntegerType(), True),
    StructField("training_name", StringType(), True),
    StructField("description", StringType(), True),
    StructField("provider", StringType(), True),
    StructField("duration_hours", IntegerType(), True)
])
```

```python
trainings_df = (
    spark.read.format("csv")
              .option("delimiter", ",")
              .schema(trainings_schema)
              .load("abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/trainings")
)
display(trainings_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
trainings_df.writeTo("zenviro.bronze.trainings_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.trainings_py;
```
