### Transformar Datos de "Photochecks"
1. Obtener el "employee_id" de "path" del archivo
2. Escribir los datos "transformados" en el schema "silver"

```python
df = spark.table("zenviro.bronze.photochecks_py")
display(df)
```

#### 1. Obtener el "employee_id" de "path" del archivo

```python
from pyspark.sql.functions import regexp_extract, col
df_extracted = (
    df
    .select(
        regexp_extract("path", r".*/([0-9]+)\.png$", 1).cast("int").alias("employee_id"),
        col("content").alias("employee_photocheck")
    )
)
display(df_extracted)
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```python
df_extracted.writeTo("zenviro.silver.photochecks_py").createOrReplace()
```
