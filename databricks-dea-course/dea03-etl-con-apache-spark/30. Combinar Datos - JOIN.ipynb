### Combinar Datos - Join(job_positions y employees)
> Unir los datos del "empleado" con su "posición" para crear una tabla "employee_position_py" que contenga la posición de cada empleado

```python
df_employees = spark.table("zenviro.silver.employees_py")
display(df_employees)
```

```python
df_job_positions = spark.table("zenviro.silver.job_positions_py")
display(df_job_positions)
```

```python
df_employee_position = (
    df_employees.join(df_job_positions,
                      df_employees.job_position_id == df_job_positions.job_position_id,
                      "inner")
    .select(
        df_employees.employee_id,
        df_employees.first_name,
        df_employees.last_name,
        df_employees.email,
        df_employees.phone,
        df_employees.status,
        df_job_positions.title,
        df_job_positions.description
    )
)
display(df_employee_position)
```

```python
df_employee_position.writeTo("zenviro.gold.employee_position_py").createOrReplace()
```
