### Leer Datos IMAGE de la carpeta "Photochecks"
1. Consultar los elementos de la carpeta "photochecks"
2. Consultar archivos "photochecks" mediante el formato BinaryFile
3. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Consultar los elementos de la carpeta "photochecks"

```python
%fs ls '/Volumes/zenviro/raw/operational_data/photochecks/'
```

#### 2. Consultar archivos "photochecks" mediante el formato BinaryFile

```python
photochecks_df = spark.read.format("binaryFile").load("/Volumes/zenviro/raw/operational_data/photochecks/*/*.png")
display(photochecks_df)
```

#### 3. Crear la Tabla(Table) en el esquema "bronze"

```python
photochecks_df.writeTo("zenviro.bronze.photochecks_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.photochecks_py;
```
