### Transformar Datos de "Employment Histories"
1. Extrae la "Fecha" de los campos "start_date" y "end_date" y darle el álias de "start_date" y "end_date"
2. Escribir los datos "transformados" en el schema "silver

```python
df = spark.table("zenviro.bronze.employment_histories_py")
display(df)
```

#### 1. Extrae la "Fecha" de los campos "start_date" y "end_date" y darle el álias de "start_date" y "end_date"

```python
from pyspark.sql.functions import date_format, col
df_extracted = (
    df
    .select(
        col("history_id").cast("int").alias("history_id"),
        col("employee_id").cast("int").alias("employee_id"),
        col("department_id").cast("int").alias("department_id"),
        col("job_position_id").cast("int").alias("job_position_id"),
        date_format("start_date", "yyyy-MM-dd").cast("date").alias("start_date"),
        date_format("end_date", "yyyy-MM-dd").cast("date").alias("end_date"),
        col("salary").cast("int").alias("salary")
    )
)
display(df_extracted)
```

#### 2. Escribir los datos "transformados" en el schema "silver

```python
df_extracted.writeTo("zenviro.silver.employment_histories_py").createOrReplace()
```
