### Leer Datos JSON de la carpeta "Eployees"
1. Consultar archivo "Employee" mediante el formato TEXT
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Consultar archivo "Employee" mediante el formato TEXT

```python
employees_df = spark.read.text("/Volumes/zenviro/raw/operational_data/employees")
display(employees_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
employees_df.writeTo("zenviro.bronze.employees_py").createOrReplace()
```
