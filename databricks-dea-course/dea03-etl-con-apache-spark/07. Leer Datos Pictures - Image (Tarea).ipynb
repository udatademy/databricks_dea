### Leer Datos IMAGE de la carpeta "Pictures"
1. Consultar los elementos de la carpeta "pictures"
2. Consultar archivos "pictures" mediante el formato BinaryFile
3. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Consultar los elementos de la carpeta "pictures"

```python
%fs ls '/Volumes/zenviro/raw/operational_data/pictures/'
```

#### 2. Consultar archivos "pictures" mediante el formato BinaryFile

```python
pictures_df = spark.read.format("binaryFile").load("/Volumes/zenviro/raw/operational_data/pictures/*/*.jpg")
display(pictures_df)
```

#### 3. Crear la Tabla(Table) en el esquema "bronze"

```python
pictures_df.writeTo("zenviro.bronze.pictures_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.pictures_py;
```
