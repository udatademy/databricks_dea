### Leer Datos TSV de la carpeta "Attendances"
1. Leer archivos "Attendances"
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Leer archivos "Attendances"

```python
attendances_df = (
    spark.read.format("csv")
              .option("header", "true")
              .option("delimiter", "\t")
              .load("/Volumes/zenviro/raw/operational_data/attendances/")
)
display(attendances_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
attendances_df.writeTo("zenviro.bronze.attendances_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.attendances_py;
```
