### Transformar Datos de "Employee Benefits"
1. Obtener solo la fecha de los campos "start_date" y "end_date"
2. Escribir los datos "transformados" en el schema "silver

```python
df = spark.table("zenviro.bronze.employee_benefits_py")
display(df)
```

#### 1. Obtener la fecha de los campos “start_date” y “end_date”

```python
from pyspark.sql.functions import date_format
df_extracted = (
    df
    .select(
        "employee_id",
        "benefit_id",
        date_format("start_date", "yyyy-MM-dd").cast("date").alias("start_date"),
        date_format("end_date", "yyyy-MM-dd").cast("date").alias("end_date")
    )
)
display(df_extracted)
```

#### 2. Escribir los datos "transformados" en el schema "silver

```python
df_extracted.writeTo("zenviro.silver.employee_benefits_py").createOrReplace()
```
