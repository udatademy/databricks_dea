### Leer Datos TSV de la carpeta "Employment History"
1. Leer todos los archivos "Employment History"
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Leer todos los archivos "Employment History"

```python
emp_hist_df = (
  spark.read.format("csv")
            .option("header", "true")
            .option("delimiter", "\t")
            .load("/Volumes/zenviro/raw/operational_data/employment_histories")
)
display(emp_hist_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
emp_hist_df.writeTo("zenviro.bronze.employment_histories_py").createOrReplace()
```
