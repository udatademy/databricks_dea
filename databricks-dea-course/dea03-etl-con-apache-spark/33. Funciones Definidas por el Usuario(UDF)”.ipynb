### Funciones Definidas por el Usuario(UDF)
- Las funciones definidas por el usuario (UDF) en Spark son funciones personalizadas creadas por los usuarios para extender las capacidades de Spark SQL y PySpark.
- Las UDF nos permiten realizar cálculos o transacciones para aplicar lógicas que no son posibles con las funciones integradas.
- Defines la función una sola vez y la utilizas en múltiples consultas.
- Se recomienda el uso de UDF de SQL en lugar de UDF de Python debido a una mejor optimización.

#### Sintaxis: <hr>
**CREATE OR REPLACE FUNCTION** catalog_name.schema_name.udf_name(param_name date_type)<br>
**RETURNS** return_type<br>
**RETURN** expression;

#### Escribe una función para concatenar el nombre y el apellido y mostrar el nombre completo.

```python
df_employees = spark.table("zenviro.silver.employees_py")
display(df_employees)
```

```python
spark.sql("""
          CREATE OR REPLACE FUNCTION zenviro.default.get_full_name_py(
            first_name STRING,
            last_name STRING
          )
          RETURNS STRING
          RETURN CONCAT(initcap(first_name), ' ', initcap(last_name))
          """)
```

```python
from pyspark.sql.functions import expr

df_result = (
    df_employees
    .select(
        "employee_id",
        expr("zenviro.default.get_full_name_py(first_name, last_name)").alias("name"),
        "hire_date",
        "email",
        "phone",
        "status"
    )
)
display(df_result)
```

```python
df_result = spark.sql("""
                        SELECT employee_id,
                            zenviro.default.get_full_name_py(first_name, last_name) AS name,
                            hire_date,
                            email,
                            phone,
                            status
                        FROM zenviro.silver.employees_py
                      """)
display(df_result)
```

```python
display(spark.sql("DESCRIBE FUNCTION EXTENDED zenviro.default.get_full_name_py"))
```

```python
def get_fullname(first_name, last_name):
    return first_name + ' ' + last_name
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

get_fullname_udf = udf(get_fullname, StringType())

df_result = (
    spark.table("zenviro.silver.employees_py")
    .select(
        "employee_id",
        get_fullname_udf("first_name", "last_name").alias("name"),
        "hire_date",
        "email",
        "phone",
        "status"
    )
)
display(df_result)
```
