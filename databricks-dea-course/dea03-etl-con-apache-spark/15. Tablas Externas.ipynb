### Tablas Externas(External Tables)
1. Crear una "Tabla Externa(External Table)" con Python
2. Describir(Desbribe) la "Tabla Externa(External Table)"
3. Efectos de eliminar una "Tabla Externa(External Table)"

#### 1. Crear una "Tabla Externa(External Table)" con Python

```python
location_df = spark.read.format("json").load("/Volumes/zenviro/raw/operational_data/locations/")
display(location_df)
```

```python
location_df.writeTo("zenviro.default.locations_ext_python") \
           .option("path", 'abfss://demo@databricksdeasa.dfs.core.windows.net/locations_ext_python') \
            .createOrReplace()
```

#### 2. Describir(Desbribe) la "Tabla Externa(External Table)"

```python
spark.sql("USE CATALOG zenviro;")
spark.sql("USE default;")
display(spark.sql("SHOW TABLES;"))
```

```python
display(spark.sql("DESCRIBE EXTENDED zenviro.default.locations_ext_python"))
```

#### 3. Efectos de eliminar una "Tabla Externa(External Table)"

```python
spark.sql("DROP TABLE IF EXISTS zenviro.default.locations_ext_python")
```
