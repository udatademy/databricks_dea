### Leer Datos JSON de la carpeta "Departments"
1. Consultar los archivos de la carpeta "Departments
2. Seleccionar metadatos del archivo
3. Registrar archivos en Unity Catalog usando Tabla(Table)

#### 1. Consultar los archivos de la carpeta "Departments"

```python
departments_df = spark.read.format("json").load("/Volumes/zenviro/raw/operational_data/departments/")
display(departments_df)
```

#### 2. Seleccionar metadatos del archivo

```python
department_with_metadata_df = departments_df.select("_metadata.file_path", "*")
display(department_with_metadata_df)
```

#### 3. Registrar archivos en Unity Catalog usando Tabla(Table)

```python
department_with_metadata_df.writeTo("zenviro.bronze.departments_py").createOrReplace()
```
