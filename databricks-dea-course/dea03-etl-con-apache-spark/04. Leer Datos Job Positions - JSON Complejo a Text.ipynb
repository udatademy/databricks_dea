### Leer Datos JSON de la carpeta "Job Positions"
1. Consultar archivo "Job Position" mediante el formato JSON
2. Consultar archivo "Job Position" mediante el formato TEXT
3. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Consultar archivo "Job Position" mediante el formato JSON

```python
job_positions_df = spark.read.json("/Volumes/zenviro/raw/operational_data/job_positions/")
display(job_positions_df)
```

#### 2. Consultar archivo "Job Position" mediante el formato TEXT

```python
job_positions_df = spark.read.text("/Volumes/zenviro/raw/operational_data/job_positions/")
display(job_positions_df)
```

#### 3. Crear la Tabla(Table) en el esquema "bronze"

```python
job_positions_df.writeTo("zenviro.bronze.job_positions_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.job_positions_py;
```
