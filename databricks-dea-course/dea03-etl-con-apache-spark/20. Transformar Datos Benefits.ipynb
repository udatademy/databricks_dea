### Transformar Datos de "Benefits"
1. Sustituir el caracter "-" por "espacio" en el campo "description"
2. Escribir los datos "transformados" en el schema "silver"

```python
df = spark.table("zenviro.bronze.benefits_py")
display(df)
```

#### 1. Sustituir el caracter "-" por "espacio" en el campo "description"

```python
from pyspark.sql.functions import regexp_replace
df_replace = (
    df
    .select(
        "benefit_id",
        "benefit_name",
        regexp_replace("description", "-", " ").alias("description"),
        "type"
    )
)
display(df_replace)
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```python
df_replace.writeTo("zenviro.silver.benefits_py").createOrReplace()
```

```python
display(spark.table("zenviro.silver.benefits_py"))
```
