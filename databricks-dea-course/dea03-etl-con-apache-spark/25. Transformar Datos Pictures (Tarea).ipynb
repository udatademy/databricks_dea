### Transformar Datos de "Pictures"
1. Obtener "location_id" del "path" del archivo
2. Escribir los datos "transformados" en el schema "silver

```python
df = spark.table("zenviro.bronze.pictures_py")
display(df)
```

#### 1. Obtener "location_id" del "path" del archivo

```python
from pyspark.sql.functions import regexp_extract, col
df_extracted = (
    df
    .select(
        regexp_extract("path", r".*/([0-9]+)\.jpg$", 1).cast("int").alias("location_id"),
        col("content").alias("location_picture")
    )
)
display(df_extracted)
```

#### 2. Escribir los datos "transformados" en el schema "silver

```python
df_extracted.writeTo("zenviro.silver.pictures_py").createOrReplace()
```
