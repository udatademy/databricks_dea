### Leer Datos PARQUET de la carpeta "Benefits"
1. Leer todos los archivos "benefits"
2. Crear la Tabla(Table) en el esquema "bronze"

#### 1. Leer todos los archivos "benefits"

```python
employee_benefits_df = (
    spark.read.format("parquet")
              .load("abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/benefits/")
)
display(employee_benefits_df)
```

#### 2. Crear la Tabla(Table) en el esquema "bronze"

```python
employee_benefits_df.writeTo("zenviro.bronze.benefits_py").createOrReplace()
```

```python
%sql
SELECT * FROM zenviro.bronze.benefits_py;
```
