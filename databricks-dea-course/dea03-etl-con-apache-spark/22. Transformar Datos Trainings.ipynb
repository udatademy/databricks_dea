### Transformar Datos de "Trainings"
1. Convertir solo la primera letra en mayúscula de los campos "training_name", "description" y "provider"
2. Escribir los datos "transformados" en el schema "silver"

```python
df = spark.table("zenviro.bronze.trainings_py")
display(df)
```

#### 1. Convertir solo la primera letra en mayúscula de los campos "training_name", "description" y "provider"

```python
from pyspark.sql.functions import initcap

df_transformed = (
    df
    .select(
        "training_id",
        initcap("training_name").alias("training_name"),
        initcap("description").alias("description"),
        initcap("provider").alias("provider"),
        "duration_hours"
    )
)
display(df_transformed)
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```python
df_transformed.writeTo("zenviro.silver.trainings_py").createOrReplace()
```
