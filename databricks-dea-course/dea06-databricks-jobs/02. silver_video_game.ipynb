### Transformar los datos de los video juegos más vendidos en la capa silver
1. Crea el esquema "**silver**" si **NO** existe.
2. Cree la tabla "**video_game**" en el esquema "**silver**" con los datos del esquema "**bronze**".
    - Crear el campo "**id**" que va a tener un valor auto-generado
    - Crear el campo "**global_sales**" que va a sumar los valores de los campos: "**na_sales**" + "**eu_sales**" + "**jp_sales**" + "**other_sales**"

```sql
CREATE SCHEMA IF NOT EXISTS demo.silver
    MANAGED LOCATION 'abfss://demo@databricksdeasa.dfs.core.windows.net/silver';
```

```sql
DROP TABLE IF EXISTS demo.silver.video_game;

CREATE TABLE demo.silver.video_game(
  id          BIGINT GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  name        STRING,
  platform    STRING,
  year        INT,
  genre       STRING,
  publisher   STRING,
  na_sales    DOUBLE,
  eu_sales    DOUBLE,
  jp_sales    DOUBLE,
  other_sales DOUBLE,
  global_sales DOUBLE GENERATED ALWAYS AS (
      ROUND(na_sales + eu_sales + jp_sales + other_sales, 2)
    )
);

INSERT INTO demo.silver.video_game 
(name, platform, year, genre, publisher, 
na_sales, eu_sales, jp_sales, other_sales)
SELECT name, platform, year, genre, publisher, 
       na_sales, eu_sales, jp_sales, other_sales
  FROM demo.bronze.video_game;    
```
