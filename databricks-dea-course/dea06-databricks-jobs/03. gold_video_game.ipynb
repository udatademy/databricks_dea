### Crear "resumen" de los video juegos mas vendidos
1. Crea el esquema "**gold**" si **NO** no existe.
2. En el esquema "**gold**" guardar la cantidad de video juegos realizado por cada editor.

```sql
CREATE SCHEMA IF NOT EXISTS demo.gold
    MANAGED LOCATION 'abfss://demo@databricksdeasa.dfs.core.windows.net/gold';
```

```sql
DROP TABLE IF EXISTS demo.gold.publisher_summary;

CREATE TABLE demo.gold.publisher_summary
AS
  SELECT publisher, count(*) AS total
  FROM demo.silver.video_game
  GROUP BY publisher
  ORDER BY total DESC;
```
