### Ingestar los datos de los video juegos más vendidos en la capa bronze
1. Crea el esquema "**bronze**" si **NO** existe.
2. Cree la tabla "**video_game**" en el esquema "**bronze**" con los datos de la carpeta "**video_games_sales**". 

```sql
CREATE SCHEMA IF NOT EXISTS demo.bronze
    MANAGED LOCATION 'abfss://demo@databricksdeasa.dfs.core.windows.net/bronze';
```

```sql
DROP TABLE IF EXISTS demo.bronze.video_game;

CREATE TABLE demo.bronze.video_game
AS
SELECT name, platform, year, genre, 
       publisher, na_sales, eu_sales, 
       jp_sales, other_sales
  FROM read_files('abfss://demo@databricksdeasa.dfs.core.windows.net/raw/video_games_sales',
                  format => 'csv',
                  delimiter => ';',
                  header => true);
```
