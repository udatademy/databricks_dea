### Transformar Datos de "Pictures"
1. Obtener "location_id" del "path" del archivo
2. Escribir los datos "transformados" en el schema "silver

```sql
SELECT * FROM zenviro.bronze.v_pictures;
```

#### 1. Obtener "location_id" del "path" del archivo

```sql
SELECT CAST(regexp_extract(path, '.*/([0-9]+)\\.jpg$', 1) AS INT) AS location_id,
       content AS location_picture
FROM zenviro.bronze.v_pictures;
```

#### 2. Escribir los datos "transformados" en el schema "silver

```sql
CREATE OR REPLACE TABLE zenviro.silver.pictures
AS
      SELECT CAST(regexp_extract(path, '.*/([0-9]+)\\.jpg$', 1) AS INT) AS location_id,
            content AS location_picture
      FROM zenviro.bronze.v_pictures;
```

```sql
SELECT * FROM zenviro.silver.pictures;
```
