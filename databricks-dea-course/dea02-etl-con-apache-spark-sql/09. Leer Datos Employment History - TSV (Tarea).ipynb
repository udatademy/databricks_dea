### Leer Datos TSV de la carpeta "Employment History"
1. Leer los archivos TSV
2. Crear la Vista(View) en el esquema "bronze"

#### 1. Leer los archivos TSV

```sql
SELECT * 
FROM read_files('/Volumes/zenviro/raw/operational_data/employment_histories',
                format => 'csv',
                delimiter => '\t',
                header => true)
```

#### 2. Crear la Vista(View) en el esquema "bronze"

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_employment_histories
AS
SELECT * 
FROM read_files('/Volumes/zenviro/raw/operational_data/employment_histories',
                format => 'csv',
                delimiter => '\t',
                header => true)
```

```sql
SELECT * FROM zenviro.bronze.v_employment_histories;
```
