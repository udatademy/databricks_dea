### Transformar Datos de "Trainings"
1. Convertir solo la primera letra en mayúscula de los campos "training_name", "description" y "provider"
2. Escribir los datos "transformados" en el schema "silver"

```sql
SELECT * FROM zenviro.bronze.trainings;
```

#### 1. Convertir solo la primera letra en mayúscula de los campos "training_name", "description" y "provider"

```sql
SELECT training_id,
        initcap(training_name) AS training_name,
        initcap(description) AS description,
        initcap(provider) AS provider,
        duration_hours
FROM zenviro.bronze.trainings;
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```sql
CREATE OR REPLACE TABLE zenviro.silver.trainings
AS
  SELECT training_id,
          initcap(training_name) AS training_name,
          initcap(description) AS description,
          initcap(provider) AS provider,
          duration_hours
  FROM zenviro.bronze.trainings;
```

```sql
SELECT * FROM zenviro.silver.trainings;
```
