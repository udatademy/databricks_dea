### Transformar Datos de "Photochecks"
1. Obtener el "employee_id" de "path" del archivo
2. Escribir los datos "transformados" en el schema "silver"

```sql
SELECT * FROM zenviro.bronze.v_photochecks
```

#### 1. Obtener el "employee_id" de "path" del archivo

```sql
SELECT CAST(regexp_extract(path, '.*/([0-9]+)\\.png$', 1) AS INT) AS employee_id,
        content AS employee_photocheck
FROM zenviro.bronze.v_photochecks;
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```sql
CREATE OR REPLACE TABLE zenviro.silver.photochecks
AS
  SELECT CAST(regexp_extract(path, '.*/([0-9]+)\\.png$', 1) AS INT) AS employee_id,
          content AS employee_photocheck
  FROM zenviro.bronze.v_photochecks;
```

```sql
SELECT * FROM zenviro.silver.photochecks;
```
