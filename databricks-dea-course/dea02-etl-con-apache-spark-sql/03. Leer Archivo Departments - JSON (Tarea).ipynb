### Leer Datos JSON de la carpeta "Departments"
1. Consultar los archivos de la carpeta "Departments
2. Seleccionar metadatos del archivo
3. Registrar archivos en Unity Catalog usando vistas(View)

#### 1. Consultar los archivos de la carpeta "Departments

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/departments/',
                        format => 'json')
```

#### 2. Seleccionar metadatos del archivo

```sql
SELECT _metadata.file_path AS file_path, *
FROM read_files('/Volumes/zenviro/raw/operational_data/departments/',
                format => 'json')
```

#### 3. Registrar archivos en Unity Catalog usando vistas(View)

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_departments
AS
SELECT _metadata.file_path AS file_path, *
FROM read_files('/Volumes/zenviro/raw/operational_data/departments/',
                format => 'json')
```

```sql
SELECT * FROM zenviro.bronze.v_departments
```
