### Resumen - Beneficio por Empleado
> Para cada "Beneficio" se necesita saber la cantidad de empleados activos durante el año "2023" y "2024"

```sql
SELECT * FROM zenviro.silver.employees;
```

```sql
SELECT * FROM zenviro.silver.benefits;
```

```sql
SELECT * FROM zenviro.silver.employee_benefits
order by start_date;
```

```sql
CREATE OR REPLACE TABLE zenviro.gold.benefits_by_employees_summary
  SELECT b.benefit_name,
        COUNT(e.employee_id) AS total_employees
  FROM zenviro.silver.benefits b
  INNER JOIN zenviro.silver.employee_benefits eb ON b.benefit_id = eb.benefit_id
  INNER JOIN zenviro.silver.employees e ON eb.employee_id = e.employee_id
  WHERE year(eb.start_date) IN (2023, 2024) AND eb.end_date IS NULL
  GROUP BY b.benefit_name
  ORDER BY COUNT(e.employee_id) DESC;
```

```sql
SELECT * FROM zenviro.gold.benefits_by_employees_summary;
```
