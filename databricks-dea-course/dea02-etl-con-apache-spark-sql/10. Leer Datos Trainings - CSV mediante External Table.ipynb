### Leer Datos CSV de la carpeta "Trainings"
1. Consultar los elementos de la carpeta "trainings"
2. Crear un "External Table"
3. Efectos de "agregar", "actualizar" o "eliminar" archivos
4. Efectos de eliminar un "External Table"

#### 1. Consultar los elementos de la carpeta "trainings"

```sql
%fs ls 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/trainings'
```

#### 2. Crear un "External Table"

```sql
CREATE TABLE IF NOT EXISTS zenviro.bronze.trainings
(
  training_id INT,
  training_name VARCHAR(50),
  description VARCHAR(255),
  provider VARCHAR(50),
  duration_hours INT
)
USING CSV
OPTIONS(
  header "false",
  delimiter = ","
)
LOCATION 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/trainings'
```

```sql
SELECT * FROM zenviro.bronze.trainings;
```

#### 3. Efectos de "agregar", "actualizar" o "eliminar" archivos

```sql
DESCRIBE EXTENDED zenviro.bronze.trainings
```

```sql
REFRESH TABLE zenviro.bronze.trainings;
```

#### 4. Efectos de eliminar un "External Table"

```sql
DROP TABLE IF EXISTS zenviro.bronze.trainings;
```
