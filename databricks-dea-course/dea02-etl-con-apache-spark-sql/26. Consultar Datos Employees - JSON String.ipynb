### Transformar Datos de "Employees" como JSON String
1. Extraer valores de la columna de nivel superior
2. Extraer elementos de la matriz(arrays)
3. Extraer valores de columnas anidadas
4. Convierte valores de columna a un tipo de dato especifico

```sql
SELECT * FROM zenviro.bronze.v_employees;
```

#### 1. Extraer valores de la columna de nivel superior<br>
> Objeto de nivel superior `<column_name>:<extraction_path>`

```sql
SELECT value:employee_id AS employee_id,
        value:department_id AS department_id,
        value
FROM zenviro.bronze.v_employees
```

#### 2. Extraer elementos de la matriz(arrays)
> Para extraer elementos de una matriz, utilice los índices entre corchetes `<column_name>:<extraction_path>[index]`

```sql
SELECT value:name[0] AS name_0,
        value:name[1] AS name_1,
        value
FROM zenviro.bronze.v_employees;
```

#### 3. Extraer valores de columnas anidadas
> Para extraer campos anidados, utilice la notación de puntos:<br> 
`<column_name>:<extraction_path>[index.<leaf node>]`

```sql
SELECT value:name[0].first_name AS first_name,
        value:name[0].last_name AS last_name,
        value
FROM zenviro.bronze.v_employees;
```

#### 4. Convierte valores de columna a un tipo de dato especifico

```sql
SELECT  value:employee_id::INTEGER AS employee_id,
        value:birth_date::DATE AS birth_date,
        value:name[0].first_name AS first_name,
        value:name[0].last_name AS last_name,
        value
FROM zenviro.bronze.v_employees;
```
