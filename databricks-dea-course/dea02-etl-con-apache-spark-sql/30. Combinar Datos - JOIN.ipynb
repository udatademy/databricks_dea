### Combinar Datos - Join(job_positions y employees)
> Unir los datos del "empleado" con su "posición" para crear una tabla "employee_position" que contenga la posición de cada empleado

```sql
SELECT * FROM zenviro.silver.employees;
```

```sql
SELECT * FROM zenviro.silver.job_positions;
```

```sql
CREATE OR REPLACE TABLE zenviro.gold.employee_position
AS
  SELECT e.employee_id,
          e.first_name,
          e.last_name,
          e.email,
          e.phone,
          e.status,
          j.title,
          j.description
  FROM zenviro.silver.employees e
  INNER JOIN zenviro.silver.job_positions j
          ON e.job_position_id = j.job_position_id
```

```sql
SELECT * FROM zenviro.gold.employee_position;
```
