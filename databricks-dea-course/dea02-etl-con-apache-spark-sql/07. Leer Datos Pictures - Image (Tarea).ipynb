### Leer Datos IMAGE de la carpeta "Pictures"
1. Consultar los elementos de la carpeta "pictures"
2. Consultar archivos "pictures" mediante el formato BinaryFile
3. Crear la Vista(View) en el esquema "bronze"

#### 1. Consultar los elementos de la carpeta "pictures"

```sql
%fs ls '/Volumes/zenviro/raw/operational_data/pictures/'
```

#### 2. Consultar archivos "pictures" mediante el formato BinaryFile

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/pictures/*/*.jpg',
                        format => 'binaryFile');
```

#### 3. Crear la Vista(View) en el esquema "bronze"

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_pictures
AS
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/pictures/*/*.jpg',
                        format => 'binaryFile');
```

```sql
SELECT * FROM zenviro.bronze.v_pictures;
```
