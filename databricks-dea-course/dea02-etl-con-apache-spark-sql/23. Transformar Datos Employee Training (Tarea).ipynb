### Transformar Datos de "Employee Training"
1. Extrae la "Fecha" y "Hora" del campo "completition_date" y crea nuevas columnas llamadas "completition_date" y "completition_time".
2. Asignar al campo "status" los siguientes valores decriptivos: <br>
   (1: Not Started, 2: In Progress, 3: Completed)
3. Escribir los datos "transformados" en el schema "silver

```sql
SELECT * FROM zenviro.bronze.employee_training;
```

#### 1. Extrae la "Fecha" y "Hora" del campo "completition_date"
[Doc - Función "date_format"](https://learn.microsoft.com/es-es/azure/databricks/sql/language-manual/functions/date_format)

```sql
SELECT employee_id,
       training_id,
       CAST(date_format(completition_date, "yyyy-MM-dd") AS DATE) AS completition_date,
       date_format(completition_date, "hh:mm:ss") AS completition_time,
       status
FROM zenviro.bronze.employee_training;
```

#### 2. Asignar al campo "status" los siguientes valores decriptivos:
##### (1: Not Started, 2: In Progress, 3: Completed)
[Doc - Función "case"](https://learn.microsoft.com/es-es/azure/databricks/sql/language-manual/functions/case)

```sql
SELECT employee_id,
       training_id,
       CAST(date_format(completition_date, "yyyy-MM-dd") AS DATE) AS completition_date,
       date_format(completition_date, "hh:mm:ss") AS completition_time,
       CASE status
        WHEN 1 THEN "Not Started"
        WHEN 2 THEN "In Progress"
        WHEN 3 THEN "Completed"
       END AS status
FROM zenviro.bronze.employee_training;
```

#### 3. Escribir los datos "transformados" en el schema "silver

```sql
CREATE OR REPLACE TABLE zenviro.silver.employee_training
AS
      SELECT employee_id,
            training_id,
            CAST(date_format(completition_date, "yyyy-MM-dd") AS DATE) AS completition_date,
            date_format(completition_date, "hh:mm:ss") AS completition_time,
            CASE status
            WHEN 1 THEN "Not Started"
            WHEN 2 THEN "In Progress"
            WHEN 3 THEN "Completed"
            END AS status
      FROM zenviro.bronze.employee_training;
```

```sql
SELECT * FROM zenviro.silver.employee_training;
```
