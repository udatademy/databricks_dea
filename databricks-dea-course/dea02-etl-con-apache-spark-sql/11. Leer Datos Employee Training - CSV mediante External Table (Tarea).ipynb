### Leer Datos CSV de la carpeta "Employee Training"
1. Consultar los elementos de la carpeta "employee_training"
2. Crear un "External Table"
3. Consultar los datos de la "External Table"

#### 1. Consultar los elementos de la carpeta "employee_training"

```sql
%fs ls 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/employee_training'
```

#### 2. Crear un "External Table"

```sql
CREATE TABLE IF NOT EXISTS zenviro.bronze.employee_training
(
  employee_id INT,
  training_id INT,
  completition_date TIMESTAMP,
  status TINYINT
)
USING CSV
OPTIONS (
  header "false",
  delimiter = ","
)
LOCATION 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/employee_training'
```

#### 3. Consultar los datos de la "External Table"

```sql
SELECT * FROM zenviro.bronze.employee_training;
```
