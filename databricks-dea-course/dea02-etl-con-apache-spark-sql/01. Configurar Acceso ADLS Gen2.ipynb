### Configurar acceso a los Datos mediante Unity Catalog
1. Crear el objeto "External Location"
2. Crear el catalog "Zenviro"
3. Crear los schemas "Raw", "Bronze", "Silver" y "Gold"
4. Crear el volumen "operational_data"

#### 1. Crear el objeto "External Location"

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricksdeasa_ext_loc_zenviro
    URL 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL storage_credential_dea)
    COMMENT 'Datos Externos del Almacenamiento Zenviro'
```

#### 2. Crear el catalog "Zenviro"

```sql
CREATE CATALOG IF NOT EXISTS zenviro
      COMMENT 'Catalog Zenviro'
```

```sql
SELECT current_catalog();
```

```sql
USE CATALOG zenviro;
```

```sql
SELECT current_catalog();
```

#### 3. Crear los schemas "Raw", "Bronze", "Silver" y "Gold"

```sql
CREATE SCHEMA IF NOT EXISTS zenviro.raw
    COMMENT 'Schema Raw';

CREATE SCHEMA IF NOT EXISTS zenviro.bronze
    COMMENT 'Schema Bronze';

CREATE SCHEMA IF NOT EXISTS zenviro.silver
    COMMENT 'Schema Silver';

CREATE SCHEMA IF NOT EXISTS zenviro.gold
    COMMENT 'Schema Gold';
```

```sql
SELECT current_schema();
```

```sql
USE SCHEMA raw;
```

```sql
SELECT current_schema();
```

#### 4. Crear el volumen "operational_data"

```sql
CREATE EXTERNAL VOLUME IF NOT EXISTS zenviro.raw.operational_data
    LOCATION 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/operational-data'
    COMMENT 'Volumen Operatinal Data';
```
