### Transformar Datos de "Employment Histories"
1. Extrae la "Fecha" de los campos "start_date" y "end_date" y darle el álias de "start_date" y "end_date"
2. Escribir los datos "transformados" en el schema "silver

```sql
SELECT * FROM zenviro.bronze.v_employment_histories;
```

#### 1. Extrae la "Fecha" de los campos "start_date" y "end_date" y darle el álias de "start_date" y "end_date"

```sql
SELECT history_id, 
       employee_id,
       department_id,
       job_position_id,
       CAST(start_date AS DATE) AS start_date,
       CAST(end_date AS DATE) AS end_date,
       salary
FROM zenviro.bronze.v_employment_histories;
```

#### 2. Escribir los datos "transformados" en el schema "silver

```sql
CREATE OR REPLACE TABLE zenviro.silver.employment_histories
AS
  SELECT history_id, 
        employee_id,
        department_id,
        job_position_id,
        CAST(start_date AS DATE) AS start_date,
        CAST(end_date AS DATE) AS end_date,
        salary
  FROM zenviro.bronze.v_employment_histories;
```

```sql
SELECT * FROM zenviro.silver.employment_histories;
```
