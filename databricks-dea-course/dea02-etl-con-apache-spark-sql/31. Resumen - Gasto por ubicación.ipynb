### Resumen - Gasto por ubicación
> Para cada "País, Ciudad y Departamento" se necesita saber los siguientes gastos:
1. Cantidad de empleados
2. Total de gasto
3. Promedio de gasto

```sql
SELECT * FROM zenviro.silver.locations;
```

```sql
SELECT * FROM zenviro.silver.departments;
```

```sql
SELECT * FROM zenviro.silver.employees;
```

```sql
SELECT * FROM zenviro.silver.employment_histories;
```

```sql
CREATE OR REPLACE TABLE zenviro.gold.employees_by_location_summary
AS
  SELECT l.country,
          l.city,
          d.department_name,
          COUNT(e.employee_id) AS total_employees,
          SUM(eh.salary) AS total_salary,
          ROUND(AVG(eh.salary), 2) AS avg_salary
  FROM zenviro.silver.locations l
  INNER JOIN zenviro.silver.departments d ON l.location_id = d.location_id
  INNER JOIN zenviro.silver.employees e ON d.department_id = e.department_id
  INNER JOIN zenviro.silver.employment_histories eh ON e.employee_id = eh.employee_id
  GROUP BY l.country, l.city, d.department_name
  ORDER BY l.country, l.city, d.department_name;
```

```sql
SELECT * FROM zenviro.gold.employees_by_location_summary;
```
