### Transformar Datos de "Attendances"
1. Extraer la "Fecha" del campo "check_in" y crear una nueva columna llamada "date"
2. Extrae la "hora" de los campos "check_in" y "check_out" y crear unas nuevas columnas llamadas "check_in_time" y "check_out_time"
3. Asignar al campo "status" los siguientes valores descriptivos:<br>
(1: Present, 2: Absent, 3: Remote, 4: Late, 5: Vacation, 6: On Leave)
4. Escribir los datos "transformados" en el schema "silver"

```sql
SELECT * FROM zenviro.bronze.v_attendances;
```

#### 1. Extraer la "Fecha" del campo "check_in" y crear una nueva columna llamada "date"

```sql
SELECT attendance_id,
        employee_id,
        CAST(check_in AS DATE) AS date,
        check_in,
        check_out,
        status
FROM zenviro.bronze.v_attendances;
```

#### 2. Extrae la "hora" de los campos "check_in" y "check_out" y crear unas nuevas columnas llamadas "check_in_time" y "check_out_time"

```sql
SELECT attendance_id,
        employee_id,
        CAST(check_in AS DATE) AS date,
        date_format(check_in, "hh:mm:ss") AS check_in_time,
        date_format(check_out, "hh:mm:ss") AS check_out_time,
        status
FROM zenviro.bronze.v_attendances;
```

#### 3. Asignar al campo "status" los siguientes valores descriptivos:<br>
(1: Present, 2: Absent, 3: Remote, 4: Late, 5: Vacation, 6: On Leave)

```sql
SELECT attendance_id,
        employee_id,
        CAST(check_in AS DATE) AS date,
        date_format(check_in, "hh:mm:ss") AS check_in_time,
        date_format(check_out, "hh:mm:ss") AS check_out_time,
        CASE status
          WHEN 1 THEN 'Present'
          WHEN 2 THEN 'Absent'
          WHEN 3 THEN 'Remote'
          WHEN 4 THEN 'Late'
          WHEN 5 THEN 'Vacation'
          WHEN 6 THEN 'On Leave'
        END AS status
FROM zenviro.bronze.v_attendances;
```

#### 4. Escribir los datos "transformados" en el schema "silver"

```sql
CREATE OR REPLACE TABLE zenviro.silver.attendances
AS
  SELECT attendance_id,
          employee_id,
          CAST(check_in AS DATE) AS date,
          date_format(check_in, "hh:mm:ss") AS check_in_time,
          date_format(check_out, "hh:mm:ss") AS check_out_time,
          CASE status
            WHEN 1 THEN 'Present'
            WHEN 2 THEN 'Absent'
            WHEN 3 THEN 'Remote'
            WHEN 4 THEN 'Late'
            WHEN 5 THEN 'Vacation'
            WHEN 6 THEN 'On Leave'
          END AS status
  FROM zenviro.bronze.v_attendances;
```

```sql
SELECT * FROM zenviro.silver.attendances;
```
