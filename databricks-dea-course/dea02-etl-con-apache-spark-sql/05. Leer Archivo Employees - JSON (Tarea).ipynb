### Leer Datos JSON de la carpeta "Employees"
1. Consultar archivo "Employee" mediante el formato TEXT
2. Crear la Vista(View) en el esquema "bronze"

#### 1. Consultar archivo "Employees" mediante el formato TEXT

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/employees',
                        format => 'TEXT')
```

#### 2. Crear la Vista(View) en el esquema "bronze"

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_employees
AS
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/employees',
                        format => 'TEXT');
```

```sql
SELECT * FROM zenviro.bronze.v_employees;
```
