### Leer Datos TSV de la carpeta "Attendances"
1. Visualizar las limitaciones del formato CSV
2. Solucionar las limitaciones del formato CSV
3. Crear la Vista(View) en el esquema "bronze"

#### 1. Visualizar las limitaciones del formato CSV

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/attendances/',
                        format => 'csv')
```

#### 2. Solucionar las limitaciones del formato CSV

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/attendances/',
                        format => 'csv',
                        delimiter => '\t',
                        header => true)
```

#### 3. Crear la Vista(View) en el esquema "bronze"

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_attendances
AS
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/attendances/',
                        format => 'csv',
                        delimiter => '\t',
                        header => true)
```

```sql
SELECT * FROM zenviro.bronze.v_attendances;
```
