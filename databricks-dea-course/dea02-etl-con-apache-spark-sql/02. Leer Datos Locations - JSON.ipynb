### Leer Datos JSON de la carpeta "Locations"
1. Consultar archivo único
2. Consultar la lista de archivos utilizando caracteres comodín
3. Consultar todos los archivos de una carpeta
4. Seleccionar metadatos de archivo
5. Registrar archivos en Unity Catalog usando Vistas(View)
6. Crear Vista Temporal(Temporary View)
7. Crear Vista Temporal Global(Global Temporary View)

```sql
%fs ls '/Volumes/zenviro/raw/operational_data/locations/'
```

#### 1. Consultar archivo único

> ##### Mediante Spark SQL

```sql
SELECT * FROM json.`/Volumes/zenviro/raw/operational_data/locations/locations_2025_10.json`
```

> ##### Mediante la función "read_files"

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/locations/locations_2025_10.json',
                        format => 'json')
```

#### 2. Consultar la lista de archivos utilizando caracteres comodín

> ##### Mediante Spark SQL

```sql
SELECT * FROM json.`/Volumes/zenviro/raw/operational_data/locations/locations_2025_*.json`
```

> ##### Mediante la función "read_files"

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/locations/locations_2025_*.json',
                        format => 'json')
```

#### 3. Consultar todos los archivos de una carpeta

> ##### Mediante Spark SQL

```sql
SELECT * FROM json.`/Volumes/zenviro/raw/operational_data/locations/`
```

> ##### Mediante la función "read_files"

```sql
SELECT * FROM read_files('/Volumes/zenviro/raw/operational_data/locations/',
                        format => 'json')
```

#### 4. Seleccionar metadatos de archivo

```sql
SELECT _metadata.file_path, * 
FROM read_files('/Volumes/zenviro/raw/operational_data/locations/',
                format => 'json')
```

#### 5. Registrar archivos en Unity Catalog usando Vistas(View)

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_locations
AS
SELECT _metadata.file_path, * 
FROM read_files('/Volumes/zenviro/raw/operational_data/locations/',
                format => 'json')
```

```sql
SELECT * FROM zenviro.bronze.v_locations;
```

#### 6. Crear Vista Temporal(Temporary View)

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_locations
AS
SELECT _metadata.file_path, * 
FROM read_files('/Volumes/zenviro/raw/operational_data/locations/',
                format => 'json')
```

```sql
SELECT * FROM tv_locations
```

#### 7. Crear Vista Temporal Global(Global Temporary View)

```sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW gtv_locations
AS
SELECT _metadata.file_path, * 
FROM read_files('/Volumes/zenviro/raw/operational_data/locations/',
                format => 'json')
```

```sql
SELECT * FROM global_temp.gtv_locations
```
