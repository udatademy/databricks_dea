### Leer Datos PARQUET de la carpeta "Benefits"
1. Consultar los elementos de la carpeta "benefits"
2. Crear un "External Table"
3. Consultar los datos de la "External Table"

#### 1. Consultar los elementos de la carpeta "benefits"

```sql
%fs ls 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/benefits'
```

#### 2. Crear un "External Table"

```sql
CREATE TABLE IF NOT EXISTS zenviro.bronze.benefits
USING PARQUET
LOCATION 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/benefits'
```

#### 3. Consultar los datos de la "External Table"

```sql
SELECT * FROM zenviro.bronze.benefits
```
