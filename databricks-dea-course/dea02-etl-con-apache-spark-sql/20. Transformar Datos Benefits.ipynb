### Transformar Datos de "Benefits"
1. Sustituir el caracter "-" por "espacio" en el campo "description"
2. Escribir los datos "transformados" en el schema "silver"

```sql
SELECT * FROM zenviro.bronze.benefits;
```

#### 1. Sustituir el caracter "-" por "espacio" en el campo "description"

```sql
SELECT benefit_id,
      benefit_name,
      REPLACE(description, "-", " ") AS description,
      type
FROM zenviro.bronze.benefits
```

#### 2. Escribir los datos "transformados" en el schema "silver"

```sql
CREATE OR REPLACE TABLE zenviro.silver.benefits
AS
  SELECT benefit_id,
        benefit_name,
        REPLACE(description, "-", " ") AS description,
        type
  FROM zenviro.bronze.benefits
```

```sql
SELECT * FROM zenviro.silver.benefits;
```
