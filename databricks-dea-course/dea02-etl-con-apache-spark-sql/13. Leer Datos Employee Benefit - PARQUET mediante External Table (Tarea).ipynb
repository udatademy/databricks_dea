### Leer Datos PARQUET de la carpeta "Employee Benefits"
1. Consultar los elementos de la carpeta "employee_benefits"
2. Crear un "External Table"
3. Consultar los datos de la "External Table"

#### 1. Consultar los elementos de la carpeta "employee_benefits"

```sql
%fs ls 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/employee_benefits'
```

#### 2. Crear un "External Table"

```sql
CREATE TABLE IF NOT EXISTS zenviro.bronze.employee_benefits
USING PARQUET
LOCATION 'abfss://zenviro@databricksdeasa.dfs.core.windows.net/raw/external-data/employee_benefits'
```

#### 3. Consultar los datos de la "External Table"

```sql
SELECT * FROM zenviro.bronze.employee_benefits;
```
