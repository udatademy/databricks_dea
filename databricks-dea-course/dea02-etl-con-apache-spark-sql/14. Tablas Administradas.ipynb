### Tablas Administradas(Managed Tables)
1. Crear una "Tabla Administrada(Managed Table)" con SQL
2. Describir(Desbribe) la "Tabla Administrada(Managed Table)"
3. Efectos de eliminar una "Tabla Administrada(Managed Table)"
4. Recuperar una "Tabla Administrada(Managed Table)"

#### 1. Crear una "Tabla Administrada(Managed Table)" con SQL

```sql
CREATE OR REPLACE TABLE zenviro.default.locations_man_sql
USING DELTA
AS
  SELECT * FROM zenviro.bronze.v_locations;
```

```sql
SELECT * FROM zenviro.default.locations_man_sql;
```

#### 2. Describir(Desbribe) la "Tabla Administrada(Managed Table)"

```sql
USE CATALOG zenviro;
USE default;
SHOW TABLES;
```

```sql
DESCRIBE EXTENDED locations_man_sql;
```

#### 3. Efectos de eliminar una "Tabla Administrada(Managed Table)"

```sql
DROP TABLE IF EXISTS zenviro.default.locations_man_sql;
```

```sql
USE CATALOG zenviro;
USE default;
SHOW TABLES;
```

#### 4. Recuperar una "Tabla Administrada(Managed Table)"

```sql
SHOW TABLES DROPPED IN zenviro.default;
```
