### Leer Datos JSON de la carpeta "Job Positions"
1. Consultar archivo "Job Position" mediante el formato JSON
2. Consultar archivo "Job Position" mediante el formato TEXT
3. Crear la Vista(View) en el esquema "bronze"

#### 1. Consultar archivo "Job Position" mediante el formato JSON

> ##### Mediante una Vista Temporal(Temporary View)

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_job_positions
USING JSON
OPTIONS(
  path "/Volumes/zenviro/raw/operational_data/job_positions/"
)
```

```sql
SELECT * FROM tv_job_positions
```

> ##### Mediante la función "read_files"

```sql
SELECT * FROM read_files("/Volumes/zenviro/raw/operational_data/job_positions/",
                        format => 'json')
```

#### 2. Consultar archivo "Job Position" mediante el formato TEXT

> ##### Mediante una Vista Temporal(Temporary View)

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_job_positions
USING TEXT
OPTIONS(
  path "/Volumes/zenviro/raw/operational_data/job_positions/"
)
```

```sql
SELECT * FROM tv_job_positions
```

> ##### Mediante la función "read_files"

```sql
SELECT * FROM read_files("/Volumes/zenviro/raw/operational_data/job_positions/",
                        format => 'text')
```

#### 3. Crear la Vista(View) en el esquema "bronze"

```sql
CREATE OR REPLACE VIEW zenviro.bronze.v_job_positions
AS
SELECT * FROM read_files("/Volumes/zenviro/raw/operational_data/job_positions/",
                        format => 'text')
```

```sql
SELECT * FROM zenviro.bronze.v_job_positions;
```
