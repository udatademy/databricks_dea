### Transformar Datos de "Employee Benefits"
1. Obtener solo la fecha de los campos "start_date" y "end_date"
2. Escribir los datos "transformados" en el schema "silver

```sql
SELECT * FROM zenviro.bronze.employee_benefits;
```

#### 1. Obtener la fecha de los campos “start_date” y “end_date”


```sql
SELECT employee_id,
       benefit_id,
       CAST(start_date AS DATE) AS start_date,
       CAST(end_date AS DATE) AS end_date
FROM zenviro.bronze.employee_benefits;
```

#### 2. Escribir los datos "transformados" en el schema "silver

```sql
CREATE OR REPLACE TABLE zenviro.silver.employee_benefits
AS
      SELECT employee_id,
            benefit_id,
            CAST(start_date AS DATE) AS start_date,
            CAST(end_date AS DATE) AS end_date
      FROM zenviro.bronze.employee_benefits;
```

```sql
SELECT * FROM zenviro.silver.employee_benefits;
```
