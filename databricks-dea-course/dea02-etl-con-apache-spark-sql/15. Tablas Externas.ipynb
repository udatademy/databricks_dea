### Tablas Externas(External Tables)
1. Crear una "Tabla Externa(External Table)" con SQL
2. Describir(Desbribe) la "Tabla Externa(External Table)"
3. Efectos de eliminar una "Tabla Externa(External Table)"

#### 1. Crear una "Tabla Externa(External Table)" con SQL

```sql
SELECT location_id, location_name, country
FROM zenviro.bronze.v_locations
```

```sql
CREATE TABLE zenviro.default.locations_ext_sql(
  location_id INT,
  location_name STRING,
  country STRING
)
USING DELTA
LOCATION "abfss://demo@databricksdeasa.dfs.core.windows.net/locations_ext_sql"
```

```sql
INSERT INTO zenviro.default.locations_ext_sql
SELECT location_id, location_name, country
FROM zenviro.bronze.v_locations
```

```sql
SELECT * FROM zenviro.default.locations_ext_sql;
```

#### 2. Describir(Desbribe) la "Tabla Externa(External Table)"

```sql
SHOW TABLES IN zenviro.default;
```

```sql
DESCRIBE EXTENDED zenviro.default.locations_ext_sql
```

#### 3. Efectos de eliminar una "Tabla Externa(External Table)"

```sql
DROP TABLE IF EXISTS zenviro.default.locations_ext_sql;
```
