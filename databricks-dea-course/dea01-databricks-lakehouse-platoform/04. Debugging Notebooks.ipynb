#### Ejemplo 1:
Explicar:
1. Breakpoints
2. Revisar código línea por línea
3. Explorador de variables
4. Consola de Depuración

```python
# Calcular el interes compuesto anual de un capital por una cantidad de años

capital = 10000         # Capital inicial
tasa_anual = 0.08       # 8% anual
anios = 1               # 12 meses

# 1. Calcular el "factor de crecimiento anual"
factor_crecimiento = 1 + tasa_anual

# 2. Elevar el factor al número de años
factor_compuesto = factor_crecimiento ** anios

# 3. Calcular el monto final
monto_final = capital * factor_compuesto

# 4. Mostrar el Monto Final
print(f"Monto Final: {monto_final:.2f}")
```

#### Ejemplo 2
Explicar:
1. Entrar (Step In) de una función
2. Salir (Step Out) de una función

```python
def interes_compuesto_anual(capital, tasa_anual, anios):
    # 1. Calcular el "factor de crecimiento anual"
    factor_crecimiento = 1 + tasa_anual

    # 2. Elevar el factor al número de años
    factor_compuesto = factor_crecimiento ** anios

    # 3. Calcular el monto final
    monto_final = capital * factor_compuesto

    return monto_final
```

```python
# Programa principal
capital = 10000         # Capital inicial
tasa_anual = 0.08       # 8% anual
anios = 1               # 12 meses

monto_final = interes_compuesto_anual(capital, tasa_anual, anios)
print(f"Monto Final: {monto_final:.2f}")
```
