### Databricks Magic Commands
- **%python, %scala, %sql, %r:** Permiten cambiar el lenguaje para una celda especifica.
- **%md:** Se utiliza para aplicar formato Markdown y documentar notebooks.
- **%fs:** Ejecutar instrucciones relacionadas con el sistema de archivos.
- **%sh:** Permite ejecutar comandos de shell (únicamente en el nodo controlador).
- **%pip:** Se emplea para instalar librerías de Python.
- **%run:** Permite incluir/importar otro Notebook en el Notebook actual.

#### %fs: Ejecutar instrucciones relacionadas con el sistema de archivos.

```python
%fs ls /databricks-datasets/
```

#### %sh: Permite ejecutar comandos de shell (únicamente en el nodo controlador).

```python
%sh ps
```

#### %pip: Se emplea para instalar librerías de Python.

```python
%pip list
```
