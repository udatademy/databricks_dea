### Transaction Log
Entendiendo la estructura de directorios del almacenamiento en la nube de las tablas "Delta Lake"

#### 1. Crear y usar un nuevo "catalog" llamado "demo"

```sql
CREATE CATALOG IF NOT EXISTS demo;
USE CATALOG demo;
```

#### 2. Crear y usar un nuevo "schema" llamado "delta_lake"

```sql
CREATE SCHEMA IF NOT EXISTS delta_lake;
USE SCHEMA delta_lake;
```

#### 3. Crear una Tabla "Delta Lake" llamada "market_ABC"

```sql
CREATE TABLE market_ABC (
  product_id INT,
  name VARCHAR(100),
  category VARCHAR(50),
  price DECIMAL(10,2),
  stock INT
);
```

```sql
DESCRIBE EXTENDED market_ABC;
```

#### 4. INSERTAR algunos datos

```sql
INSERT INTO market_ABC (product_id, name, category, price, stock)
VALUES (1, 'Laptop Pro 15', 'Tecnología', 4500.00, 10);
```
