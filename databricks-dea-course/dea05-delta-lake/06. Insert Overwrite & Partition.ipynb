### "Insert Overwrite" & "Partition"

### Insert Overwrite
1. Reemplazar todos los datos de una tabla
2. Reemplazar todos los datos de una partición específica

> INSERT OVERWRITE -> Sobrescribe los datos existentes en una tabla o una partición específica con los nuevos datos.<br>
> INSERT INTO -> Añade nuevos datos.

> INSERT OVERWRITE -> Utilícelo para sobrescribir los datos de una tabla(sin columnas generadas) o partición cuando no haya cambios en el esquema.<br>
> CREATE OR REPLACE TABLE -> Utilizar cuando haya cambios en el esquema.

#### 1. Reemplazar todos los datos de una tabla

```sql
DROP TABLE IF EXISTS demo.delta_lake.gold_sales;

CREATE TABLE demo.delta_lake.gold_sales
(
  sales_id LONG NOT NULL,
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE,
  category STRING
);

INSERT INTO demo.delta_lake.gold_sales
(sales_id, product_name, quantity, price, total, category)
VALUES
(1, 'Laptop', 2, 1200.50, 2401, 'Technology'),
(2, 'Keyboard Mechanical', 5, 95.99, 479.95, 'Technology');

SELECT * FROM demo.delta_lake.gold_sales;
```

```sql
DROP TABLE IF EXISTS demo.delta_lake.bronze_sales;

CREATE TABLE demo.delta_lake.bronze_sales
(
  sales_id LONG NOT NULL,
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE,
  category STRING
);

INSERT INTO demo.delta_lake.bronze_sales
(sales_id, product_name, quantity, price, total, category)
VALUES
(1, 'Laptop', 2, 1200.50, 2401, 'Technology'),
(2, 'Keyboard Mechanical', 5, 95.99, 479.95, 'Technology'),
(3, 'Wireless Mouse', 3, 45.75, 137.25, 'Technology'), 
(4, 'External SSD 1TB', 4, 210.25, 841, 'Technology'),
(5, 'Office Desk', 1, 1850.00, 1850, 'Furniture'),
(6, 'Office Sofa', 1, 3900.00, 3900, 'Furniture');

SELECT * FROM demo.delta_lake.bronze_sales;
```

```sql
INSERT OVERWRITE TABLE demo.delta_lake.gold_sales
SELECT *
FROM demo.delta_lake.bronze_sales;
```

```sql
SELECT * FROM demo.delta_lake.gold_sales;
```

```sql
DESCRIBE HISTORY demo.delta_lake.gold_sales;
```

#### 2. Reemplazar todos los datos de una partición específica

```sql
DROP TABLE IF EXISTS demo.delta_lake.gold_sales_partitioned;

CREATE TABLE demo.delta_lake.gold_sales_partitioned
(
  sales_id LONG NOT NULL,
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE,
  category STRING
)
PARTITIONED BY (category);

INSERT INTO demo.delta_lake.gold_sales_partitioned
(sales_id, product_name, quantity, price, total, category)
VALUES
(1, 'Laptop', 2, 1200.50, 2401, 'Technology'),
(2, 'Keyboard Mechanical', 5, 95.99, 479.95, 'Technology');

SELECT * FROM demo.delta_lake.gold_sales_partitioned;
```

```sql
DESCRIBE EXTENDED demo.delta_lake.gold_sales_partitioned;
```

```sql
DROP TABLE IF EXISTS demo.delta_lake.bronze_sales_technology;

CREATE TABLE demo.delta_lake.bronze_sales_technology
(
  sales_id LONG NOT NULL,
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE,
  category STRING
)
PARTITIONED BY (category);

INSERT INTO demo.delta_lake.bronze_sales_technology
(sales_id, product_name, quantity, price, total, category)
VALUES
(1, 'Laptop', 2, 1200.50, 2401, 'Technology'),
(2, 'Keyboard Mechanical', 5, 95.99, 479.95, 'Technology'),
(3, 'Wireless Mouse', 3, 45.75, 137.25, 'Technology'), 
(4, 'External SSD 1TB', 4, 210.25, 841, 'Technology');
```

```sql
SELECT * FROM demo.delta_lake.bronze_sales_technology
```

```sql
INSERT OVERWRITE TABLE demo.delta_lake.gold_sales_partitioned
PARTITION (category = 'Technology')
SELECT sales_id,
       product_name,
       quantity,
       price,
       total
FROM demo.delta_lake.bronze_sales_technology
```

```sql
SELECT * FROM demo.delta_lake.gold_sales_partitioned;
```

```sql
DROP TABLE IF EXISTS demo.delta_lake.bronze_sales_furniture;

CREATE TABLE demo.delta_lake.bronze_sales_furniture
(
  sales_id LONG NOT NULL,
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE,
  category STRING
);

INSERT INTO demo.delta_lake.bronze_sales_furniture
(sales_id, product_name, quantity, price, total, category)
VALUES
(5, 'Office Desk', 1, 1850.00, 1850, 'Furniture'),
(6, 'Office Sofa', 1, 3900.00, 3900, 'Furniture');
```

```sql
SELECT * FROM demo.delta_lake.bronze_sales_furniture;
```

```sql
INSERT OVERWRITE TABLE demo.delta_lake.gold_sales_partitioned
PARTITION (category = 'Furniture')
SELECT sales_id,
       product_name,
       quantity,
       price,
       total
FROM demo.delta_lake.bronze_sales_furniture;
```

```sql
SELECT * FROM demo.delta_lake.gold_sales_partitioned;
```
