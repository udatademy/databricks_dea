### "Create or Replace" vs "Create Table AS (CTAS)"
1. Diferencia entre las sentencias "Create or Replace" y "Drop y Create Table"
2. Sentencia "Create Table AS (CTAS)"

#### 1. Diferencia entre las sentencias "Create or Replace" y "Drop y Create Table"

##### Sentencias "Drop" y "Create"

```sql
DROP TABLE IF EXISTS demo.delta_lake.sales;
CREATE TABLE demo.delta_lake.sales
(
  sales_id LONG NOT NULL GENERATED ALWAYS AS IDENTITY ( START WITH 1 INCREMENT BY 1),
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE GENERATED ALWAYS AS (quantity * price)
);

INSERT INTO demo.delta_lake.sales
(product_name, quantity, price)
VALUES
('Laptop', 2, 1200.50),
('Keyboard Mechanical', 5, 95.99),
('Wireless Mouse', 3, 45.75),
('External SSD 1TB', 4, 210.25);
```

```sql
DESCRIBE HISTORY demo.delta_lake.sales;
```

##### Sentencia "Create or Replace"

```sql
DROP TABLE IF EXISTS demo.delta_lake.sales;
```

```sql
CREATE OR REPLACE TABLE demo.delta_lake.sales
(
  sales_id LONG NOT NULL GENERATED ALWAYS AS IDENTITY ( START WITH 1 INCREMENT BY 1),
  product_name STRING,
  quantity INT,
  price DOUBLE,
  total DOUBLE GENERATED ALWAYS AS (quantity * price)
);

INSERT INTO demo.delta_lake.sales
(product_name, quantity, price)
VALUES
('Laptop', 2, 1200.50),
('Keyboard Mechanical', 5, 95.99),
('Wireless Mouse', 3, 45.75),
('External SSD 1TB', 4, 210.25);
```

```sql
DESCRIBE HISTORY demo.delta_lake.sales;
```

#### 2. Sentencia "Create Table As (CTAS)"

```sql
DROP TABLE demo.delta_lake.sales_low_ticket;
```

```sql
CREATE TABLE demo.delta_lake.sales_low_ticket
AS
  SELECT CAST(sales_id AS INT) AS sales_id,
         product_name,
         quantity,
         price,
         total
  FROM demo.delta_lake.sales
  WHERE total < 500;
```

```sql
SELECT * FROM demo.delta_lake.sales_low_ticket;
```

```sql
DESCRIBE HISTORY demo.delta_lake.sales_low_ticket;
```

```sql
DESCRIBE EXTENDED demo.delta_lake.sales_low_ticket;
```

```sql
ALTER TABLE demo.delta_lake.sales_low_ticket
  ALTER COLUMN price COMMENT "El precio del producto."
```

```sql
ALTER TABLE demo.delta_lake.sales_low_ticket
  ALTER COLUMN sales_id SET NOT NULL;
```

```sql
ALTER TABLE demo.delta_lake.sales_low_ticket
  ALTER COLUMN product_name SET NOT NULL;
```
