### Table & Column Properties
Demostrar las propiedades de tabla y columna que podemos agregar mediante la instrucción CREATE TABLE.

#### 1. Table Properties
- 1.1. COMMENT - Permiten documentar el propósito de la tabla.
- 1.2. TBLPROPERTIES - Utilizado para especificar metadatos a nivel de tabla o ajustes de configuración.

```sql
DROP TABLE IF EXISTS demo.delta_lake.sales;
CREATE TABLE demo.delta_lake.sales
(
  sales_id LONG,
  product_name STRING,
  quantity INT,
  price DOUBLE
)
COMMENT "Esta tabla contiene información de la empresa."
TBLPROPERTIES(
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true,
  delta.logRetentionDuration = '30 days',
  delta.deletedFileRetentionDuration = '7 days',
  delta.appendOnly = true
)
;
```

```sql
DESCRIBE EXTENDED demo.delta_lake.sales;
```

```sql
SHOW TBLPROPERTIES demo.delta_lake.sales;
```

#### 2. Column Properties
- 2.1. NOT NULL(Constraints) - Garantiza la integridad y la calidad de los datos al asegurar que una columna específica no pueda contener valores NULL.
- 2.2. COMMENT - Documenta el propósito o contexto de las columnas individuales de una tabla.
- 2.3. Generated Columns - Son Columnas derivadas o calculadas, cuyos valores se calculan en el momento de insertar nuevos registros.
    - 2.3.1. Generated Identity Columns - Utilizado para generar una identidad, por ejemplo, un valor de clave principal.
    - 2.3.2. Generated Computed Columns - Calcular y almacenar automáticamente valores derivados basados en otras columnas de la misma tabla.

##### 2.3.1 Generated Identity Columns
`GENERATED { ALWAYS | BY DEFAULT } AS IDENTITY [ ( [ START WITH start ] [INCREMENT BY step] ) ]`

```sql
DROP TABLE IF EXISTS demo.delta_lake.sales;
CREATE TABLE demo.delta_lake.sales
(
  sales_id LONG NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  product_name STRING NOT NULL,
  quantity INT NOT NULL,
  price DOUBLE COMMENT "Contiene el precio del producto."
)
COMMENT "Esta tabla contiene información de la empresa."
TBLPROPERTIES(
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true,
  delta.logRetentionDuration = '30 days',
  delta.deletedFileRetentionDuration = '7 days',
  delta.appendOnly = true
)
;
```

```sql
INSERT INTO demo.delta_lake.sales
(product_name, quantity, price)
VALUES
('Laptop', 2, 1200.50),
('Keyboard Mechanical', 5, 95.99),
('Wireless Mouse', 3, 45.75),
('External SSD 1TB', 4, 210.25);
```

```sql
SELECT * FROM demo.delta_lake.sales;
```

```sql
DESCRIBE EXTENDED demo.delta_lake.sales;
```

##### 2.3.2. Generated Computed Columns
GENERATED ALWAYS AS (`exp`)

`exp` puede estar compuesto por literales, identificadores de columna dentro de la tabla y funciones u operadores SQL deterministas e integrados, excepto:
- Aggregate functions (Funciones de Agregación)
- Analytics windows functions (Funciones de Ventanas Analíticas)
- Ranking window functions (Funciones de Ventana de Clasificación)
- Table valued generator functions (Funciones Generadoras de Valores de Tabla)

Además `exp` no debe contener ninguna subconsulta.

```sql
DROP TABLE IF EXISTS demo.delta_lake.sales;
CREATE TABLE demo.delta_lake.sales
(
  sales_id LONG NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  product_name STRING NOT NULL,
  quantity INT NOT NULL,
  price DOUBLE COMMENT "Contiene el precio del producto.",
  total DOUBLE GENERATED ALWAYS AS (quantity * price)
)
COMMENT "Esta tabla contiene información de la empresa."
TBLPROPERTIES(
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true,
  delta.logRetentionDuration = '30 days',
  delta.deletedFileRetentionDuration = '7 days',
  delta.appendOnly = true
)
;
```

```sql
INSERT INTO demo.delta_lake.sales
(product_name, quantity, price)
VALUES
('Laptop', 2, 1200.50),
('Keyboard Mechanical', 5, 95.99),
('Wireless Mouse', 3, 45.75),
('External SSD 1TB', 4, 210.25);
```

```sql
SELECT * FROM demo.delta_lake.sales;
```

#### 3. Table Properties
- 3.1. **...UNSET...** - Permite eliminar completamente una propiedad de la tabla.
- 3.2. **...SET...** - Permite actualizar el valor de una propiedad existente (o crearla si no existe).

```sql
DELETE FROM demo.delta_lake.sales
WHERE sales_id = 1;
```

```sql
ALTER TABLE demo.delta_lake.sales
UNSET TBLPROPERTIES(
  delta.appendOnly
)
```

```sql
SELECT * FROM demo.delta_lake.sales;
```

```sql
ALTER TABLE demo.delta_lake.sales
SET TBLPROPERTIES(
  delta.logRetentionDuration = '20 days'
)
```

```sql
SHOW TBLPROPERTIES demo.delta_lake.sales;
```
