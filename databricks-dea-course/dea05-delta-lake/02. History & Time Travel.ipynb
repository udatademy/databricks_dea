### History & Time Travel
1. Consultar el historial de la tabla Delta Lake.
2. Consultar versiones anteriores de los datos.
3. Consultar datos de un momento específico.
4. Restaurar datos a una versión específica.

#### 1. Consultar el historial de la tabla Delta Lake.

```sql
DESCRIBE HISTORY demo.delta_lake.market_abc;
```

#### 2. Consultar versiones anteriores de los datos.

```sql
SELECT * FROM demo.delta_lake.market_abc;
```

```sql
SELECT * FROM demo.delta_lake.market_abc
VERSION AS OF 1;
```

#### 3. Consultar datos de un momento específico.

```sql
SELECT * FROM demo.delta_lake.market_abc
TIMESTAMP AS OF '2026-04-10T22:54:46.000+00:00'
```

#### 4. Restaurar datos a una versión específica.

```sql
RESTORE TABLE demo.delta_lake.market_abc VERSION AS OF 1;
```

```sql
SELECT * FROM demo.delta_lake.market_abc;
```
