### Sentencia "Copy Into"
> - Carga datos de forma incremental en tablas Delta Lake desde Cloud Storage.
> - Admite la evolución del esquema.
> - Admite una amplia gama de formatos de archivo (CSV, JSON, Parquet, Delta).
> - Alternativa a "Auto Loader" para la ingesta por "batch".

Documentación - [COPY INTO](https://learn.microsoft.com/es-es/azure/databricks/sql/language-manual/delta-copy-into)

#### Crear la tabla para copiar los datos en ella.

```sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.raw_products_inventory;
```

#### Cargar nuevos archivos en la tabla de forma incremental.

```sql
DELETE FROM demo.delta_lake.raw_products_inventory;

COPY INTO demo.delta_lake.raw_products_inventory
FROM 'abfss://demo@databricksdeasa.dfs.core.windows.net/raw/products_inventory'
FILEFORMAT = JSON
FORMAT_OPTIONS ('inferSchema' = 'true')
COPY_OPTIONS ('mergeSchema' = 'true')
```

```sql
SELECT * FROM demo.delta_lake.raw_products_inventory;
```
