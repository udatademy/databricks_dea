### Compactación mediante OPTIMIZE
Combina varios archivos pequeños en unos pocos archivos grandes, mejorando así el rendimiento.
> 1. Crear la tabla vacia "**demo.delta_lake.order**"
> 2. Inserta datos como transacción individual
> 3. Consultar el historial de la Tabla
> 4. Número de archivos leidos para obtener los Datos
> 5. Ejecutar "OPTIMIZE"

#### 1. Crear la tabla vacia "**demo.delta_lake.order**"

```sql
DROP TABLE IF EXISTS demo.delta_lake.order;

CREATE TABLE IF NOT EXISTS demo.delta_lake.order
(
  order_id STRING,
  city STRING,
  order_date DATE,
  weight DOUBLE
)
```

#### 2. Inserta datos como transacción individual

```sql
INSERT INTO demo.delta_lake.order
VALUES('ODR_00001', 'New York', '2026-01-10', 9.2981);
```

```sql
INSERT INTO demo.delta_lake.order
VALUES('ODR_00002', 'New York', '2026-01-11', 7.6808);
```

```sql
INSERT INTO demo.delta_lake.order
VALUES('ODR_00003', 'San Diego', '2026-01-12', 1.7123);
```

```sql
INSERT INTO demo.delta_lake.order
VALUES('ODR_00004', 'Dallas', '2026-01-13', 2.4925);
```

#### 3. Consultar el historial de la Tabla

```sql
DESC HISTORY demo.delta_lake.order;
```

#### 4. Número de archivos leidos para obtener los Datos

```sql
DESCRIBE DETAIL demo.delta_lake.order;
```

```sql
SELECT * FROM demo.delta_lake.order;
```

#### 5. Ejecutar "OPTIMIZE"

```sql
OPTIMIZE demo.delta_lake.order;
```

```sql
SELECT * FROM demo.delta_lake.order VERSION AS OF 3;
```
