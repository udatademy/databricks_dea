### VACUUM
> Elimina de forma permanente archivos obsoletos de una tabla Delta para recuperar espacio en almacenamiento. Tiene un periodo de retención de 7 días (168 horas) por defecto.

#### 1. Comprueba el historial de la Tabla Delta.

```sql
DESCRIBE HISTORY demo.delta_lake.market_abc;
```

#### 2. Ejecutar la instrucción VACUUM

```sql
SET spark.databricks.delta.retentionDurationCheck.enabled = false;
VACUUM demo.delta_lake.market_abc RETAIN 0 HOURS;
```

```sql
SELECT * FROM demo.delta_lake.market_abc
VERSION AS OF 3;
```
