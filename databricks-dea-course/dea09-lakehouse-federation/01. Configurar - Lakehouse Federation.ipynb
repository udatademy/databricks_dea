#### Configurar acceso a Azure SQL mediante Lakehouse Federation
1. Crear Conexión - sql_zenviro_db_conn_sql
2. Crear Catalog Externo - asql_zenviro_db_catalog_sql

```sql
CREATE CONNECTION sql_zenviro_db_conn_sql TYPE sqlserver
OPTIONS (
  host 'zenviro-srv.database.windows.net',
  port '1433',
  user 'adminsql',
  password 'Password@1'
);
```

```sql
CREATE FOREIGN CATALOG IF NOT EXISTS asql_zenviro_db_catalog_sql USING CONNECTION sql_zenviro_db_conn_sql
OPTIONS (database 'zenviro-db');
```

```sql
SHOW CATALOGS;
USE CATALOG asql_zenviro_db_catalog_sql;
SHOW SCHEMAS;
USE SCHEMA dbo;
SHOW TABLES;
SELECT * FROM benefits;
```
