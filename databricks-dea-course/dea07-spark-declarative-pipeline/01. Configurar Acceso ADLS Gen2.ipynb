### Configuración del entorno Tixo - Data Lakehouse
1. Crear "External Location" - databricksdeasa_ext_loc_tixo
2. Crear el "Catalog" - tixo
3. Crear los "Schemas"
    - raw
    - bronze
    - silver
    - gold
4. Crear el "Volume" - operational_data

#### 1. Crear "External Location"
**Nombre:** databricksdeasa_ext_loc_tixo<br>
**Path ADLS:** [abfss://tixo@databricksdeasa.dfs.core.windows.net/](abfss://tixo@databricksdeasa.dfs.core.windows.net/)<br>
**Storage Credential:** storage_credential_dea

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricksdeasa_ext_loc_tixo
  URL 'abfss://tixo@databricksdeasa.dfs.core.windows.net/'
  WITH (STORAGE CREDENTIAL storage_credential_dea)
  COMMENT 'Datos Externos del Almacenamiento Tixo';
```

```sql
%fs ls 'abfss://tixo@databricksdeasa.dfs.core.windows.net/'
```

#### 2. Crear el "Catalog"
**Nombre:** tixo<br>
**Managed Location:** [abfss://tixo@databricksdeasa.dfs.core.windows.net/](abfss://tixo@databricksdeasa.dfs.core.windows.net/)

```sql
CREATE CATALOG IF NOT EXISTS tixo
  MANAGED LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/'
  COMMENT 'Catalog Tixo';
```

```sql
SHOW CATALOGS;
```

#### 3. Crear los "Schemas"
**1. Nombre:** raw <br>
&emsp;_Managed Location:_ [abfss://tixo@databricksdeasa.dfs.core.windows.net/raw](#)<br>
**2. Nombre:** bronze <br>
&emsp;_Managed Location:_ [abfss://tixo@databricksdeasa.dfs.core.windows.net/bronze](#)<br>
**3. Nombre:** silver <br>
&emsp;_Managed Location:_ [abfss://tixo@databricksdeasa.dfs.core.windows.net/silver](#)<br>
**4. Nombre:** gold <br>
&emsp;_Managed Location:_ [abfss://tixo@databricksdeasa.dfs.core.windows.net/gold](#)

```sql
USE CATALOG tixo;

CREATE SCHEMA IF NOT EXISTS raw
  MANAGED LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/raw';

CREATE SCHEMA IF NOT EXISTS bronze
  MANAGED LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/bronze';

CREATE SCHEMA IF NOT EXISTS silver
  MANAGED LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/silver';

  CREATE SCHEMA IF NOT EXISTS gold
  MANAGED LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/gold';
```

```sql
SHOW SCHEMAS;
```

#### 4. Crear el "Volume"
**Nombre:** operational_data<br>
**Path ADLS:** [abfss://tixo@databricksdeasa.dfs.core.windows.net/raw/operational-data](#)

```sql
USE CATALOG tixo;
USE SCHEMA raw;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
  LOCATION 'abfss://tixo@databricksdeasa.dfs.core.windows.net/raw/operational-data'
```

```sql
%fs ls /Volumes/tixo/raw/operational_data
```
